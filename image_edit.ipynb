```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img
```

```python
# read image
image = img.imread('./figures/ling_red.jpg')
```

```python
# RGB
print(f'the shape of ling image: {image.shape}')
```

```python
# 0 - 255 int, sometimes float
red = image[:, :, 0]
green = image[:, :, 1]
blue = image[:, :, 2]

print(red.shape, green.shape, blue.shape)
```

```python
print(red)
```

```python
plt.imshow(image, interpolation='nearest')
plt.axis('off');
```

```python
color_title = ['red', 'green', 'blue']
fig, ax = plt.subplots(1, 3, figsize=(12, 4))
for i, axi in enumerate(ax):
    im = axi.imshow(image[:, :, i], cmap='jet', interpolation='nearest')    
    axi.set_title(color_title[i], size = 16)
    fig.colorbar(im, ax=axi)
```

```python
new_img = np.copy(image)
# red
new_img[:, :, 0] = blue
# green
new_img[:, :, 1] = green
# blue
new_img[:, :, 2] = red

plt.imshow(new_img,interpolation='nearest')
plt.axis('off');
```

```python
new_img = np.copy(image)
new_img[new_img[:, :, 0] > 234] = 255

plt.imshow(new_img)
plt.axis('off');

# save image
plt.savefig('./figures/new_img.jpg')
```
